```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading census dataset
census = pd.read_csv('CensusDataset.csv')
census.drop(columns=['fnlwgt'], inplace=True)
census.head()
```

```python
target = 'class'
census[target].value_counts()
```

```python
census.info()
```

```python
# Numerical and Categorical columns
cat = census.dtypes == object
num = census.dtypes != object
cat_col = census.loc[:,cat].columns
num_col = census.loc[:,num].columns
```

```python
print(f'Shape of dataset: {census.shape[0]} x {census.shape[1]}')
```

```python
# Histogram for numerical values
_ = census.hist(figsize=(20,14))
```

```python
# Distribution of some categorical values
print('Sex:\n',census.sex.value_counts(),'\n')
print('Education:\n',census.education.value_counts())
```

```python
# Check relationship between education and education num
pd.crosstab(index=census.education, columns=census['education-num'])
```

```python
# Plot only a subset
samples = 5000
columns = ['age', 'education-num', 'hours-per-week']
_ = sns.pairplot(data = census[:samples],
                vars = columns,
                hue = target,
                plot_kws = {'alpha' : 0.2},
                height = 3,
                diag_kind = 'hist',
                diag_kws = {'bins' : 30})
```

```python
#Ploting boundries manually
ax = sns.scatterplot(x = 'age',
                   y = 'hours-per-week',
                   data = census[:samples],
                   hue = 'class',
                   alpha = 0.5)

age = 27
plt.axvline(x=age, ymin=0, ymax=1, color='black', linestyle='--')

hours = 40
plt.axhline(y=hours, xmin=0.18, xmax=1, color='black', linestyle='--')

plt.annotate('<= 50K', (17,25), rotation=90, fontsize=35)
plt.annotate('<= 50K', (35,20), fontsize=35)
_ = plt.annotate('???', (45,60), fontsize=35)
```
